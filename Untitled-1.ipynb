```python
import scipy.io
import pandas as pd
import numpy as np

# List of file paths
file_paths = [
    "C:/Users/moham/Resarch project/data_preprocessed/data_preprocessed/Data_Preprocessed_P01.mat"]


# Initialize an empty list to store dataframes from each file
dataframes = []

for file_path in file_paths:
    # Load the .mat file
    data = scipy.io.loadmat(file_path)

    # Initialize an empty dictionary to store reshaped arrays for this file
    reshaped_data = {}

    # Iterate through the dictionary and reshape each array
    for key, value in data.items():
        if isinstance(value, (list, np.ndarray)):
            if isinstance(value, list):
                # If it's a list, convert it to a NumPy array
                value = np.array(value)
            if len(value.shape) > 1:
                # Reshape to a 1D array or a shape that suits your data
                reshaped_data[key] = value.flatten()

    # Create a DataFrame from the reshaped data for this file
    df = pd.DataFrame(reshaped_data)

    # Append the DataFrame to the list
    dataframes.append(df)

# Concatenate all dataframes into one
final_df1 = pd.concat(dataframes, ignore_index=True)


final_df1.head()
```
